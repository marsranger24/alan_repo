```python
def shift(n, N): 
    s = np.zeros(N) 
    s[n] = 1.0 
    return np.fft.fft(s)  

def calc_phase_difference(fftA, fftB, chA_delay, chB_delay,max_bin_number=1000):

    phaseA = np.angle(fftA[:max_bin_number]) - chA_delay
    phaseB = np.angle(fftB[:max_bin_number]) - chB_delay

    diff = np.array(phaseA - phaseB)
    while diff[diff<-pi].any():
        diff[diff<-pi] = diff[diff<-pi] + 2.*pi
    while diff[diff>pi].any():
        diff[diff>pi] = diff[diff>pi] - 2.*pi

    return diff #np.array(phaseA - phaseB)

def reshape2d( dat, bins=8192 ):
    ch_2d = np.reshape(dat,[-1,bins])
    ch_2d = np.flip(ch_2d,1)
    ch_2d = np.rot90(ch_2d)
    return( ch_2d )


ymax = 1000
xmax = 250

fft_orig1 = proc.inputdata.fft_ch4#[:,0]
fft_orig2 = proc.inputdata.fft_ch2#[:,0]
print(fft_orig1.shape)
S = shift(fft_orig1.shape[0]//2-1, fft_orig1.shape[0])
# S[:, None]
S_tile = np.tile(S,fft_orig1.shape[1])
S_tile = reshape2d(S_tile)
fft_shift1 = fft_orig1*S_tile
# fft_shift2 = fft_orig2*S[:, None]
print(fft_shift1.shape)

phase_orig = calc_phase_difference(fft_orig1,fft_orig2,0,0,8192)
phase_shift = calc_phase_difference(fft_shift1, fft_orig2,0,0,8192)

# fig, ax = plt.subplots()
#what you were doing:
#figuring out the right way to apply S

fig, ax = plt.subplots()
im = plt.imshow(np.real(S_tile[0:ymax,0:xmax]), aspect='auto',origin='lower',cmap=plt.cm.seismic, vmin=-1, vmax=1)


fig, ax = plt.subplots()
im = plt.imshow(np.log10(np.abs(fft_shift1[0:ymax,0:xmax])), aspect='auto',origin='lower',cmap=plt.cm.viridis, vmin=5, vmax=10)

fig, ax = plt.subplots()
im = plt.imshow(phase_orig[0:ymax,0:xmax], aspect='auto',origin='lower',cmap=plt.cm.seismic, vmin=-pi, vmax=pi)

fig, ax = plt.subplots()
im = plt.imshow(phase_shift[0:ymax,0:xmax], aspect='auto',origin='lower',cmap=plt.cm.seismic, vmin=-pi, vmax=pi)
```

```python
import numpy as np
from matplotlib.pyplot import plot, legend
def shift(n, N): 
    s = np.zeros(N) 
    s[n] = 1.0 
    return np.fft.fft(s)  
t = np.linspace(0, 2*np.pi,1000) 
v0 = np.sin(t)                                                                                               
S = shift(1000//4, 1000)  # shift by pi/4
V = np.fft.fft(v0)
VS = np.fft.fft(v0)*S 
vs = np.fft.ifft(VS)
fig, ax = plt.subplots()
plot(t, v0 , label='original' )
plot(t,vs.real,label='shifted in frequency domain')
legend()

fig, ax = plt.subplots()
plot(V.real,label='vreal')
plot(V.imag,label='vimag')
plot(S.real,label='sreal')
plot(S.imag,label='simag')
plot(VS.real,label='vsreal')
plot(VS.imag,label='vsimag')
legend()
```
